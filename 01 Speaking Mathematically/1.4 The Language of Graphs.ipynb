### Description of a Graph

- A graph consists of a set of vertices and a set of edges connecting various pairs of vertices. The edges may be straight or curved and should either connect one vertex to another or a vertex to itself.

```python
import graphviz
```

```python
# Create Graph object
dot = graphviz.Graph()
```

```python
# Add nodes
dot.node('1' ,'start')
dot.node('2')
dot.node('3')
dot.node('4')
dot.node('5')
```

```python
# Add edge between 1 and 2
dot.edge('1' ,'2', 'e1', constraint='true')

# Add edge between 1 and 3
dot.edges(['13'])

# Add edge between 3 and 4
dot.edges(['34'])

# Add edge between 3 and 5
dot.edge('3' ,'5', constraint='false')
```

```python
# Visualize the graph
dot
```

### loop - an edge with just one endpoint

```python
# Add edge between 2 and 2
dot.edges(['22'])
```

```python
# Visualize the graph
dot
```

### parallel - two or more distinct edges with the same set of endpoints

```python
# Add edge between 5 and 3
dot.edge('5' ,'3')
```

```python
# Visualize the graph
dot
```

### A Mathematical description of a Graph

- A graph $G$ consists of two finite sets: a nonempty set $V$ of vertices and a set $E$ of edges, where each edge is associated with a set consisting of either one or two vertices called its endpoints.

### Definition of a Graph

A graph is an edge-endpoint function.

### Directed Graph

- A graph $G$ consists of two finite sets: a nonempty set $V$ of vertices and a set $D_E$ of directed edges, where each edge is associated with an ordered pair vertices called its endpoints.

- RECAP - An ordered pair means the first element belongs to let's say a set A and the second element below to let's say a set B, such that there's a relation between elements of A and elements of B

```python
# Create Digraph object
dot = graphviz.Digraph()
```

```python
# Add nodes
dot.node('1','Milky Way')
dot.node('2','Solar System')
dot.node('3','Sun')
dot.node('4','Earth')
dot.node('5','Mars')
```

```python
# Add edge between 1 and 1
dot.edge('1' ,'2')

# Add edge between 2 and 3
dot.edge('2' ,'3')

# Add edge between 3 and 4
dot.edge('3' ,'4')

# Add edge between 3 and 5
dot.edge('3' ,'5')
```

```python
# Visualize the graph
dot
```

### Degree of a Vertex - count of edges incident on a vertex

- $deg(\text{Milky Way})$ is 1
- $deg(\text{Solar System})$ is 2
- $deg(\text{Sun})$ is 3
- $deg(\text{Earth})$ is 1
